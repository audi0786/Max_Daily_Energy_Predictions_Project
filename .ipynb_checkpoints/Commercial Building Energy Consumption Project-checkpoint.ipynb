```python
price_demand_df = pd.read_csv("price_demand_data.csv", sep = ',', header = 0)
```

```python
price_demand_df
```

```python
price_demand_df.info()
```

```python
price_demand_df['SETTLEMENTDATE'] = pd.to_datetime(price_demand_df.SETTLEMENTDATE)
price_demand_df['SETTLEMENTDATE_date']=price_demand_df.SETTLEMENTDATE.dt.date
price_demand_df
```

```python
pd_df = price_demand_df.groupby('SETTLEMENTDATE_day').agg(max=('TOTALDEMAND','max'), mean=('TOTALDEMAND','mean'), median=('TOTALDEMAND','median'), count=('TOTALDEMAND', 'count'))
pd_df = pd_df.reset_index()
pd_df
```
